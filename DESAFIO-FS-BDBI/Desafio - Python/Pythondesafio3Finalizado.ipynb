# Agrupamento de Dados e Plotagem de Gráficos

### Pesquisa: Quantidade de funcionários em cada setor

```python
QtdFuncionariosSetor = df['Department'].value_counts()
QtdFuncionariosSetor.plot(kind='bar')

plt.title('Quantidade de Funcionários por Setor')
plt.xlabel('Setor')
plt.ylabel('Quantidade de Funcionários')
```

### Pesquisa: Quantidade de funcionários para cada nível de escolaridade

```python
QtdFuncionariosEsc = df['Education'].value_counts()
QtdFuncionariosEsc.plot(kind='line', marker='o')

plt.title('Quantidade de Funcionários por Escolaridade')
plt.xlabel('Nível de Escolaridade')
plt.ylabel('Quantidade de Funcionários')

plt.xticks(rotation=45)
plt.show()
```

### Pesquisa: Quantidade de funcionários que tem ou não algum atrito na empresa

```python
QtdFuncionariosAtt = df['Attrition'].value_counts()
QtdFuncionariosAtt.plot(kind='pie', autopct='%1.1f%%')

plt.title('Quantidade de Funcionários que tem Atrito ou não na empresa')
plt.ylabel('')
plt.show()
```
