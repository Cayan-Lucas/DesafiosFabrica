# Agrupamento de Dados e Plotagem de Gr√°ficos

### Pesquisa: Quantidade de funcion√°rios em cada setor

```python
QtdFuncionariosSetor = df['Department'].value_counts()
QtdFuncionariosSetor.plot(kind='bar')

plt.title('Quantidade de Funcion√°rios por Setor')
plt.xlabel('Setor')
plt.ylabel('Quantidade de Funcion√°rios')
```

### Pesquisa: Quantidade de funcion√°rios para cada n√≠vel de escolaridade

```python
QtdFuncionariosEsc = df['Education'].value_counts()
QtdFuncionariosEsc.plot(kind='line', marker='o')

plt.title('Quantidade de Funcion√°rios por Escolaridade')
plt.xlabel('N√≠vel de Escolaridade')
plt.ylabel('Quantidade de Funcion√°rios')

plt.xticks(rotation=45)
plt.show()
```

### Pesquisa: Quantidade de funcion√°rios que tem ou n√£o algum atrito na empresa

```python
QtdFuncionariosAtt = df['Attrition'].value_counts()
QtdFuncionariosAtt.plot(kind='pie', autopct='%1.1f%%')

plt.title('Quantidade de Funcion√°rios que tem Atrito ou n√£o na empresa')
plt.ylabel('')
plt.show()
```
